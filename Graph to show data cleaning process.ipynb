```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
df_d = pd.read_csv('/Users/sarahcliff/Desktop/hydrology data/raw data/dbh2raw@2022-02-17.csv')
df_c = pd.read_csv('/Users/sarahcliff/Desktop/hydrology data/deseasoned data/dbh2deseasoned@2022-02-15.csv')
df_b = pd.read_csv('/Users/sarahcliff/Desktop/hydrology data/detrended data/dbh2detrended@2022-02-13.csv')
df_a = pd.read_csv('/Users/sarahcliff/Desktop/hydrology data/autoreg hydrology data/dbh2autoreg@2022-02-13.csv')
```

```python
date_d = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in df_d['Date']]
data_d = df_d['Data']
date_c = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in df_c['Date']]
data_c = df_c['Data']
date_b = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in df_b['Date']]
data_b = df_b['Data']
date_a = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in df_a['Date']]
data_a = df_a['Data']
```

```python
#plotting everything
csfont = {'fontname':'Times New Roman'}
left, width = .25, .5
bottom, height = .25, .5
top = bottom + height
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
axs[1,1].plot(date_a, data_a)
axs[1,1].set_title('A', loc = 'center', **csfont)
axs[1,0].plot(date_b, data_b)
axs[1,0].set_title('B', loc = 'center', **csfont)
axs[0,1].plot(date_c, data_c)
axs[0,1].set_title('C', loc = 'center', **csfont)
axs[0,0].plot(date_d, data_d)
axs[0,0].set_title('D', loc = 'center', **csfont)

fig.suptitle('Data Processing Rounds for DBH2', **csfont, fontsize =16)
fig.text( y=0.5,x=0.05,s='Depth in Centimeters', rotation = 'vertical',**csfont, fontsize = 16 )
fig.text( y=0.05,x=0.5,s='Date', rotation = 'horizontal',**csfont, fontsize = 16 )
```
